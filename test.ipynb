```python
from mss import mss
import cv2
import numpy as np
import pyautogui
from PIL import Image
```

```python
width, height = pyautogui.size()
top = 0
left  = int(width * 0.4)
width = int(width * 0.4)
print(width, height)
```

```python
left_note = cv2.imread("notes//left_note.png", 0)
lw, lh = left_note.shape[::-1]
right_note = cv2.imread("notes//right_note.png", 0)
rw, rh = right_note.shape[::-1]
left_wide_note = cv2.imread("notes//left_wide_note.png", 0)
lww, lwh = left_wide_note.shape[::-1]
right_wide_note = cv2.imread("notes//right_wide_note.png", 0)
rww, rwh = right_wide_note.shape[::-1]
```

```python
left_gray = cv2.cvtColor(left_note, cv2.COLOR_BGR2GRAY)
right_gray = cv2.cvtColor(right_note, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("left_note", left_note)
```

```python
with mss() as sct:
    monitor = {'top': top, 'left': left, 'width': width, 'height': height}
    #screenshot = sct.grab(monitor)
    screenshot = cv2.imread("screenshot.png")
    img = np.array(screenshot)

    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    res_l = cv2.matchTemplate(img_gray, left_note, cv2.TM_CCOEFF_NORMED)
    res_r = cv2.matchTemplate(img_gray, right_note, cv2.TM_CCOEFF_NORMED)
    
    threshold = 0.8
    loc_l = np.where(res_l >= threshold)
    loc_r = np.where(res_r >= threshold)
    
    for pt in zip(*loc_l[::-1]):
        cv2.rectangle(img, pt, (pt[0] + lw, pt[1] + lh), (0, 255, 0), 2)
    
    for pt in zip(*loc_r[::-1]):
        cv2.rectangle(img, pt, (pt[0] + rw, pt[1] + rh), (0, 255, 0), 2)
    
```

```python
cv2.imwrite("output.png", img)
```
